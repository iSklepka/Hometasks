Добавим описание столбцов: 

- school - школа ученика (двоичный: ‘GP’ - Gabriel Pereira или ‘MS’ - Mousinho da Silveira)

- sex - пол ученика (двоичный: ‘F’ - женский or ‘M’ - мужской)

- age - возраст ученика (числовой: от 15 до 22)

- address - тип домашнего адресса ученика (двоичный: ‘U’ - город or ‘R’ - село)

- famsize - размер семьи (двоичный: ‘LE3’ - меньше или равно 3 or ‘GT3’ - больше чем 3)

- Pstatus - статс совместного проживания родителей (двоичный: ‘T’ - вместе or ‘A’ - раздельно)

- Medu - образование матери (числовой: 0 - нет, 1 - начальное образование (4 класса), 2 - с 5 по 9 классы, 3 - среднее образование или 4 — высшее образование)

- Fedu - образование отца (числовой: 0 - нет, 1 - начальное образование (4 класса), 2 - с 5 по 9 классы, 3 - среднее образование или 4 — высшее образование)

- Mjob - работа матери (номинальный: «учитель», «связанное со здравоохранением», гражданские «службы» (например, административные или полицейские), «на дому» или «другое»)

- Fjob - работа отца (номинальный: «учитель», «связанное со здравоохранением», гражданские «службы» (например, административная или полицейская), «на дому» или «другое»)

- reason - причина выбора школы (номинальный: близкая к «дому», «репутация» школы, предпочтение «курса» или «другое»)

- guardian - опекун студента (номинальный: «мать», «отец» или «другой»)

- traveltime - время в пути от дома до школы (числовой: 1 – <15 минут, 2 – 15 – 30 минут, 3 – 30 минут до 1 часа или 4 – >1 часа)

- studytime - еженедельное время обучения (числовой: 1 — <2 часов, 2 — 2–5 часов, 3 — 5–10 часов или 4 — >10 часов)

- failures - количество прошлых ошибок класса (числовой: n, если 1<=n<3, иначе 4)

- schoolsup - дополнительная образовательная поддержка (двоичный: да или нет)

- famsup - образовательная поддержка семьи (двоичный: да или нет)

- paid - дополнительные платные занятия по предмету курса (португальский) (двоичный: да или нет)

- activities - внеклассные мероприятия (бинарный: да или нет)

- nursery - посещал детский сад (двоичный: да или нет)

- higher - хочет получить высшее образование (двоичный: да или нет)

- Internet — доступ в Интернет дома (двоичный: да или нет)

- romantic – с романтическими отношениями (бинарный: да или нет)

- famrel - качество семейных отношений (числовой: от 1 - очень плохо до 5 - отлично)

- freetime – свободное время после школы (числовой: от 1 – очень низкое до 5 – очень высокое)

- goout – встреча с друзьями (числовой: от 1 – очень низкий до 5 – очень высокий)

- Dalc - потребление алкоголя в течение рабочего дня (числовой: от 1 - очень низкое до 5 - очень высокое)

- Walc – потребление алкоголя в выходные дни (числовой: от 1 – очень низкое до 5 – очень высокое)

- health – текущее состояние здоровья (числовой: от 1 – очень плохое до 5 – очень хорошее)

- absences - количество пропусков в школе (числовой: от 0 до 93)

- G1 – оценка первого урока (числовой: от 0 до 20)

- G2 – оценка за второй период (числовой: от 0 до 20)

- G3 — итоговая оценка (числовой: от 0 до 20, выходная цель)

```python
# Присвоим категориальных переменных - числовое значение
new_df = pd.get_dummies(df, columns=['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet',
 'romantic'])
```

```python
# Отделим целевую переменную от всего массива
X = new_df.drop('G3', axis=1).values
y = new_df['G3'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_test
```

```python
# Масштабирование данных с помощью pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('std_scalar', StandardScaler())
])

X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
X_test
```

# Моделирование

```python
# Линейная регрессия

lin_reg = LinearRegression()

lin_reg.fit(X_train,y_train)
```

```python
# Добавим метрики

from sklearn import metrics
from sklearn.model_selection import cross_val_score

def cross_val(model):
    pred = cross_val_score(model, X, y, cv=10)
    return pred.mean()

def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
    
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

```python
test_pred = lin_reg.predict(X_test)
train_pred = lin_reg.predict(X_train)

print('Test set evaluation:\n_____________________________________')
print_evaluate(y_test, test_pred)
print('Train set evaluation:\n_____________________________________')
print_evaluate(y_train, train_pred)
```

```python
results_df = pd.DataFrame(data=[["Linear Regression", *evaluate(y_test, test_pred) , cross_val(LinearRegression())]], 
                          columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Square', "Cross Validation"])
results_df
```
